## Question 5

```python
#  function used in Flask for URL Building - url_for()
import flask
app=flask.Flask(__name__)

@app.route('/')
def home():
    return 'This is Home'

@app.route('/temp')
def temp():
    return 'This is temporary function'

@app.route('/<name>')
def temp1(name):
    if name=='temp':
        return flask.redirect(flask.url_for('temp'))

if __name__=='__main__':
    app.run(host='0.0.0.0')
```
