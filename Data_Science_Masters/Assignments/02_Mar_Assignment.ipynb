## Question 2

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.show()
```

## Question 3

```python
# plt.subplots()-Create a figure and a set of subplots.This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
ax1 = plt.subplot(2,2,1)
ax1.plot(x1,y1)
ax2 = plt.subplot(2,2,2)
ax2.plot(x2,y2)
ax3 = plt.subplot(2,2,3)
ax3.plot(x2,y3)
ax4 = plt.subplot(2,2,4)
ax4.plot(x4,y4)
```

## Question 4

```python
# bar plot - is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

## Question 5

```python
# Box Plot-a simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles, usually with a vertical line inside to indicate the median value. The lower and upper quartiles are shown as horizontal lines either side of the rectangle.
# it is used to detect outliers
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.show()
plt.boxplot(box2)
plt.show()
```
